### Question 1. Write a function called sum-of-squares that determines the sum of the squares of two numbers. 

<pre>
(format t "Write a function called Sum of Squares that determines the sum of the squares of two numbers.~%")

(defun Sum-of-Squares (x y)
  (+ (* x x) (* y y)))

(format t "Enter the first number(x): ")
(let ((x (read)))
  (format t "Enter the second number(y): ")
  (let ((y (read)))
    (format t "Sum of Squares (x*x + y*y) = ~a" (Sum-of-Squares x y))))
</pre>

### Question 2: Write a function called greater that determines the greater number out of two numbers.

<pre>
(format t "Write a function called greater that determines the greater number out of two numbers.~%")

(defun greater (a b)
  (if (> a b)
      a
      b))

(format t "Enter the first number(x): ")
(let ((x (read)))
  (format t "Enter the second number(y): ")
  (let ((y (read)))
    (format t "Greater Number = ~a" (greater x y))))
</pre>

### Question 3: Write a function called fifth that detects the fifth number in an array. Also your program should say NIL in case your array contains less than 5 numbers.

<pre>
(format t "Write a function called fifth that detects the fifth number in an array. Also your program should say NIL in case your array contains less than 5 numbers.~%")

(defun my-fifth (arr)
    (if (>= (length arr) 5)
        (nth 4 (coerce arr 'list))
        NIL))

(format t "Please enter an array with at least five numbers, like #(1 2 3 4 5):  ")
(let ((arr (read)))
    (format t "The fifth number in the array is: = ~a~%" (my-fifth arr)))

</pre>
